```python
import pandas as pd
```

```python
utube=pd.read_csv(r"D:\DATA SCIENCE\DATASET\youtube_ad_revenue.csv")
```

```python
utube.duplicated().sum()
```

```python
utube.drop_duplicates(inplace=True)
```

```python
utube.isnull().sum()
```

```python
utube['likes']=utube['likes'].fillna(utube['likes'].mean())
utube['comments']=utube['comments'].fillna(utube['comments'].mean())
utube['watch_time_minutes']=utube['watch_time_minutes'].fillna(utube['watch_time_minutes'].mean())
```

```python
utube.count()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in utube.select_dtypes(include=["object"]).columns:
    utube[col] = le.fit_transform(utube[col])
```
